**Data Exploration**
The minimum page number is zero which logically no book with zero page. Thus, we will take a look at books with less than 5 pages.

```python
books = books.rename(columns={"# num_pages": "num_pages"})
books_min = books[(books.num_pages < 5)]
books_min
```

The books with less than 5 pages actually have ratings. There might be some errors when inputting the page number. However some of them only have a few ratings and reviews count. We will take a deeper look to these books.


```python
# This will be the group of books that need to be checked.

books_min = books_min[(books_min.ratings_count < 10) & (books_min.text_reviews_count < 5)]
books_min
```

Then we will explore about book ratings. First we will see how many 0 rating books with at least one review.

```python
books_zero = books[(books.average_rating == 0) & (books.text_reviews_count > 0)]
books_zero
```

Then, books that have average rating but no rating counts.

```python
books_rates = books[(books.average_rating > 0) & (books.ratings_count < 1)]
books_rates
```

Next, we will explore the data and plot some charts to see the feature according to book language and number of pages. Firstly, we will use groupby and nunique to show any language listed in the dataset and how many books for each group.

```python
books_lang = books.groupby('language_code')['bookID'].nunique()
books_lang
```

All different English type will be regroup as one, and languages except international language will be grouped as others.

```python
books_p = books.replace({'en-CA':'eng','en-US':'eng','en-GB':'eng'})
books_p['language_code'] = books_p['language_code'].replace(
                           ['ale','ara','cat','dan','enm','gla','glg',
                           'grc','heb','lat','msa','nl','nor','rus','srp','swe','tur','wel'] , 'othr')
```

```python
books_lang = books_p.groupby('language_code')['bookID'].nunique()
books_lang
```

Due to huge amount of English books, we will plot a pie chart to show the ratio of English books and non-English books. Then, another pie chart also plotted to show the ratio of non-English books.

```python
labels = 'English', 'Others'
sizes = [12643, 1055]
colors = ['royalblue', 'orange']

# Plot
plt.pie(sizes, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
